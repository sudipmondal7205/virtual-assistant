```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('Crop_Yield_Prediction.csv')
```

```python
df1.sample(6)
```

```python
df2 = pd.read_csv("Crop(Distric level).csv")
```

```python
df2['label'].value_counts()
```

```python
df1
```

```python
df1.columns = df1.columns.str.replace('Nitrogen', 'N')
df1.columns = df1.columns.str.replace('Phosphorus', 'P')
df1.columns = df1.columns.str.replace('Potassium', 'K')
df1.columns = df1.columns.str.replace('pH_Value', 'ph')
```

```python
df1
```

```python
df2.columns = df2.columns.str.replace('temperature', 'Temperature')
df2.columns = df2.columns.str.replace('humidity', 'Humidity')
df2.columns = df2.columns.str.replace('rainfall', 'Rainfall')   
df2.columns = df2.columns.str.replace('label', 'Crop')
```

```python
df2
```

```python
df1 = df1[['N', 'P', 'K', 'ph', 'Rainfall', 'Temperature', 'Humidity', 'Crop']]
df2 = df2[['N', 'P', 'K', 'ph', 'Rainfall', 'Temperature', 'Humidity', 'Crop']]
```

```python
df1
```

```python
df2
```

```python
df = pd.concat([df1, df2], axis=0, ignore_index=True)
```

```python
df.sample(6)
```

```python
df.to_csv('crop-prediction.csv', index=False)
```
